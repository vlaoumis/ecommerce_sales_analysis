```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "OrderID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Customer": ["John", "Maria", "John", "Anna", "Nick", "Maria", "George", "Anna"],
    "Product": ["Laptop", "Phone", "Mouse", "Keyboard", "Phone", "Laptop", "Mouse", "Phone"],
    "Category": ["Electronics", "Electronics", "Accessories", "Accessories", "Electronics", "Electronics", "Accessories", "Electronics"],
    "Quantity": [1, 2, 3, 1, 1, 1, 2, 1],
    "UnitPrice": [800, 500, 25, 70, 500, 800, 25, 500],
    "OrderDate": ["2026-01-05", "2026-01-10", "2026-02-02", "2026-02-12", "2026-03-01", "2026-03-15", "2026-03-20", "2026-04-01"]
}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python

df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
df
```

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
df.info()
```

```python
df["Month"] = df["OrderDate"].dt.to_period("M")
df
```

```python
total_revenue = df["Revenue"].sum()
print(f"Total revenue: €{total_revenue}")
```

```python
top_products = df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
top_products
```

```python
top_products.plot(kind="bar")
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue (€)")
plt.show()
```

```python
revenue_by_category = df.groupby("Category")["Revenue"].sum().sort_values(ascending=False)
revenue_by_category
```

```python
revenue_by_category.plot(kind="bar")
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue (€)")
plt.show()
```

```python
monthly_revenue = df.groupby("Month")["Revenue"].sum()
monthly_revenue
```

```python
monthly_revenue.plot(kind="line", marker="o")
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue (€)")
plt.show()
```

```python
average_order_value = df["Revenue"].mean()
print(f"Average Order Value: €{average_order_value:.2f}")
```

```python
orders_per_customer = df.groupby("Customer")["OrderID"].count().sort_values(ascending=False)
orders_per_customer
```

```python
df.to_csv("cleaned_ecommerce_sales.csv", index=False)
```
